```python
list_countries=["youtube-new/CA_category_id.json","youtube-new/US_category_id.json","youtube-new/GB_category_id.json"]
```

```python
for country in list_countries:
    json_data=json.load(open(country, 'r'))
    for data in json_data['items']:
        cat.append([data['id'],country[12:14],data['snippet']['title']])
            
cat
```

```python
category_df=pd.DataFrame(cat,columns=['Category_id','Country','Category_name'])
category_df.head(200)
```
